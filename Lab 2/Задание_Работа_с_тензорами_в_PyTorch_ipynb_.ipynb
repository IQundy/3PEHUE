Задание 17. Выполните стыковку тензоров a и b по вертикали.

Вход:

> a = torch.arange(2, 10, 2).reshape(2, -1)

> b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $4 \times 5$.


```python
# ваш код здесь
a = torch.arange(2, 10, 2).reshape(2, -1)
b = torch.ones(10).reshape(2, -1)
a1 = a.repeat(1, 3)[:, :5]
vniz = torch.cat([a1, b], dim = 0)
print(vniz)
```

Задание 18. Выполните стыковку тензоров a и b по горизонтали.

Вход:

a = torch.arange(2, 10, 2).reshape(2, -1)

b = torch.repeat(1, 10).reshape(2, -1)

Желаемый результат: новый тензор размером $2 \times 10$.

```python
# ваш код здесь
a = torch.arange(2, 10, 2).reshape(2, -1)
b = torch.ones(10).reshape(2, -1)
vpravo = torch.cat([a, b], dim = 1)
print(vpravo)
```

Задание 19. Получите все элементы от 5 до 10 в тензоре a.

Вход:

> a = torch.tensor([2, 6, 1, 9, 10, 3, 27])

Желаемый результат:

> tensor([6, 9, 10])

```python
# ваш код здесь
a = torch.tensor([2, 6, 1, 9, 10, 3, 27])
mask = (5 <= a) & (a <= 10)
otr = a[mask]
print(otr)
```

Задание 23. Вставить в двумерный терзор из задания 17 строку из минимального элемента перед первой строкой.

```python
# ваш код здесь
min = vniz.min()
row = torch.full((1, vniz.shape[1]), min)
otvet = torch.cat([row, vniz], dim=0)
print(otvet)
```
